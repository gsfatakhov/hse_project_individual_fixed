```python
from Bio import SeqIO
```

```python
!mkdir scaffolds
```

```python
input_fasta = 'GCF_000313135.1_Acastellanii.strNEFF_v1_genomic.fna'
```

```python
output_dir = 'scaffolds/'


def split_fasta_by_scaffolds(input_fasta, output_dir):
    fasta_sequences = SeqIO.parse(open(input_fasta), 'fasta')
    for fasta in fasta_sequences:
        name, sequence = fasta.id, str(fasta.seq)
        with open(output_dir + name + ".fasta", "w") as output_file:
            output_file.write(f">{name}\n{sequence}\n")


split_fasta_by_scaffolds(input_fasta, output_dir)
```

```python
%%bash
DIRECTORY="./scaffolds/"

cd $DIRECTORY

for file in *; do
    if [ -f "$file" ]; then
        ../zhunt3 12 8 12 "$file"
        echo "Processed: $file"  # Выводим сообщение о обработанном файле
    fi
done
```

```python
import pandas as pd
import os
```

```python
res = []

for file in os.listdir("scaffolds"):
    if file.endswith(".fasta.Z-SCORE"):
        filename = file.rstrip('.fasta.Z-SCORE')

        df = pd.read_table(os.path.join("scaffolds", file),
                           sep='\s+',
                           skiprows=1,
                           names=['Start', 'End', 'None1', 'None2', 'None3', 'Z-score', 'Sequence', 'None4'])

        df = df[df['Z-score'] >= 400]

        for index, row in df.iterrows():
            res.append((filename, row['Start'], row['End']))
```

```python
print(len(res))
print(res[:10])
```

```python
zhunt_df = pd.DataFrame(res, columns=['Chromosome', 'Start', 'End'])
```

#### Начинаем работу со скаффолдами

```python
import re
```

```python
pqs = []
pqs_minus = []

template = "(?:G{3,}[ATGC]{1,7}){3,}G{3,}"
template_minus = "(?:C{3,}[ATGC]{1,7}){3,}C{3,}"

for rec in SeqIO.parse(input_fasta, "fasta"):
    name, seq = rec.id, str(rec.seq)
    pqs += [[name, m.start(), m.end(), m.group(0), '+'] for m in
            re.finditer(template, seq, re.IGNORECASE)]
    pqs_minus += [[name, m.start(), m.end(), m.group(0), '-'] for m in
                  re.finditer(template_minus, seq, re.IGNORECASE)]
```

```python
print(pqs[:100])
```

```python
print(len(pqs))
print(len(pqs_minus))
```

Получаем экзоны

```python
!awk '{if($1 !~ /^#/ ) print $1"\t"$4-1"\t"$5"\t"$3"\t"0"\t"$7}'  GCF_000313135.1_Acastellanii.strNEFF_v1_genomic.gtf > annotation.bed
```

```python
!awk '$4 == "exon"' annotation.bed > exons.bed
```

Получаем квадруплексы

```python
with open('./quadrs.bed', 'a') as f:
    for record in pqs + pqs_minus:
        f.write(record[0] + "\t" + str(record[1]) + "\t" + str(record[2]) + "\tdummy\tdummy\t" + record[4] + "\n")
```

Пересекаем квадруплексы с экзонами

```python
!bedtools intersect -a quadrs.bed -b exons.bed > inters_quadrs_exons.bed
```

```python
!wc -l inters_quadrs_exons.bed
```

Получаем интроны

```python
!awk '$4=="transcript"' annotation.bed > transcripts.bed
```

```python
!bedtools subtract -a transcripts.bed -b exons.bed > introns.bed
```

Пересекаем квадруплексы с интронами

```python
!bedtools intersect -a quadrs.bed -b introns.bed > inters_quadrs_itrons.bed
```

```python
!wc -l inters_quadrs_itrons.bed
```

Получаем размеры скаффолдов и гены

```python
with open('scaf_sizes.txt', 'w') as out:
    for record in SeqIO.parse('GCF_000313135.1_Acastellanii.strNEFF_v1_genomic.fna', "fasta"):
        scaffold_name = record.id
        scaffold_length = len(record.seq)
        out.write(f"{scaffold_name}\t{scaffold_length}\n")
```

```python
!awk 'BEGIN{OFS="\t"} {if ($4 == "gene") {print $1, $2, $3, $4, $5, $6}}' annotation.bed > genes.bed
```

Получаем промоутеры

```python
!bedtools flank -i genes.bed -g scaf_sizes.txt -l 1000 -r 0 -s > promoters.bed
```

```python
!bedtools intersect -a quadrs.bed -b promoters.bed -s > inters_quadrs_promoters.bed
```
